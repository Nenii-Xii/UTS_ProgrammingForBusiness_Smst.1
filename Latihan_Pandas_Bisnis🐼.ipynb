```python
import pandas as pd

#Membaca dataset penjualan
df = pd.read_excel('Penjualan Retail.xlsx')

#Hitung total keseluruhan pendapatan perusahaan
df['Pendapatan'] = df['Harga'] * df['Jumlah']
total_pendapatan = df['Pendapatan'].sum()
print("1. Total Pendapatan Perusahaan:")
print(f"Rp {total_pendapatan:,.0f}\n")
```

```python
#Tampilkan produk dengan harga tertinggi
produk_termahal = df.loc[df['Harga'].idxmax(), ['Produk', 'Harga']]
print("2. Produk dengan Harga Tertinggi:")
print(produk_termahal, "\n")
```

```python

#Buat kolom baru Biaya = 20% dari (Harga x Jumlah)
df['Biaya'] = 0.2 * df['Pendapatan']
```

```python
#Buat kolom Laba = Pendapatan - Biaya
df['Laba'] = df['Pendapatan'] - df['Biaya']
```

```python
#Tampilkan produk dengan Laba terbesar
produk_Laba_terbesar = df.loc[df['Laba'].idxmax(), ['Produk', 'Laba']]
print("5. Produk dengan Laba Terbesar:")
print(produk_laba_terbesar, "\n")
```
